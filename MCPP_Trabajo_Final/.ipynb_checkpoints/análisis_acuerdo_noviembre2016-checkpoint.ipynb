```python
import numpy as np
import pandas as pd
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [15.0, 8.0]
import re
import string
```

```python
with open('acuerdo_paz_nov.txt') as f:
    acuerdo2 = f.read()
```

```python
print (acuerdo2)
```

### 1. Limpiando el texto

```python
# Hacer una lista con todos los caractéres
all_tokens = acuerdo2.split()
all_tokens
```

```python
# Definir los caracteres de puntuación
punctuation = string.punctuation + '–¡¿”“•\r´'
punctuation
```

```python
# Quitar todos los caractéres de puntuación de la lista, definiendo una función para este fin
def no_punct(string):
    transtable = string.maketrans('', '', punctuation)
    return string.translate(transtable)
tokens_no_punc = list(map(no_punct, all_tokens))
tokens_no_punc
```

```python
# Dejar en la lista sólo los caractéres alfabéticos
alpha = [w for w in tokens_no_punc if not w.isdigit()]
alpha
```

```python
# pasar todas las aplabras a minúsculas
lower = [w.lower() for w in alpha]
lower
```

```python
# Importar las "stopwords" en español
from nltk.corpus import stopwords
stopwords_esp = stopwords.words('spanish')
stopwords_esp
```

```python
# Filtrar el texto de las "stopwords" importadas
tokens_no_stop = [w for w in lower if w not in stopwords_esp]
tokens_no_stop
```

```python
print (sorted(set(tokens_no_stop)))
```

```python
# Se realiza una primera extracción de palabras que están mal escritas en el texto
extract1 = ['', '10º', '1definir', '1que', '1º', '2definir', '2desplazamiento', '2º', '2ºdicho', '353a', '356a', '3con', '3fin', '3que', '3º', '4para', '4quienes', '4º', '5adoptar', '5º', '6a', '6º', '7para', '7º', '8definir', '8º', '9recibir', '9º', 'a1', 'aarak', '\ufeffacuerdo', 'q', 'qr', 'san', 'x', 'x1', 'xx', 'xxx', 'xxxx', 'xxxxxx','—', '—hombres', '—mujeres', '—niñas', '—niños', '—que', '…', '\ufeffacuerdo']
extracting1 = [w for w in tokens_no_stop if w not in extract1]
extracting1
```

```python
# Imprimir una lista de frecuencias para detectar palabras que no son de interés en el análisis del texto
freqdist = nltk.FreqDist(extracting1)
frec = freqdist.most_common()
print(frec)
```

```python
# Sacar palabras que se detectaron como de no interés para el análisis del texto
no_interest = ['página','mmv','así','ser','da','cfhbd','caso','uso','punto','ptn','general','más','g','p','cuenta','será','artículo','respecto','objeto','todas','cimmv','materia','dentro','vez','hechos','demás','según','mismo','casos','forma','base','si','podrán','jep','ello','bajo','ep','u','d','c','pnis','misma','trata','hará','anexo','toda','está','hace','estén','i','evento','hacen','cuentan','habrá','h','trate','cuya','contar','cuyo','d180','cabo','aquí','salvo','d5','hagan','sólo','qué','que','j','aei','menor','kroc','cuáles','d90','todosas','veces','sigue','xxx','xxxxxx']
extracting2 = [w for w in extracting1 if w not in no_interest]
print(sorted(extracting2))
```

## 2. Hacer el wordcloud

```python
# Buscar las 50 palabras más importantes para hacer wordcloud en https://tagul.com/
freqdist2 = nltk.FreqDist(extracting2)
frec2 = freqdist2.most_common(50)
frec2
```
